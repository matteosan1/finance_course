```python
import pandas as pd

mq = pd.read_csv("input_files/ois_quotes.csv")
print (mq.head())
```

```python
from finmarkets import generate_swap_dates, OvernightIndexSwap, DiscountCurve
from datetime import date

observation_date = date(2019, 10, 23)
pillar_dates = [observation_date]
swaps = []
print (len(mq))
for i in range(len(mq)):
    swap = OvernightIndexSwap(1e6,
                              generate_swap_dates(observation_date,
                                  mq['maturity'].tolist()[i]),
                              0.01 * mq['rate'].tolist()[i])
    swaps.append(swap)
    pillar_dates.append(swap.payment_dates[-1])
pillar_dates = sorted(pillar_dates)
print (len(pillar_dates))
```

```python
import numpy as np
obj_val = []
def objective_function(x, obs_date, pillars):
    x = np.insert(x, 0, 1)
    curve = DiscountCurve(obs_date, pillars, x)
    sum_sq = 0.0
    for swap in swaps:
        sum_sq += swap.npv(curve) ** 2
    return sum_sq

def pippo(x):
    x = np.insert(x, 0, 1)
    curve = DiscountCurve(observation_date, pillar_dates, x)
    sum_sq = 0.0
    for swap in swaps:
        sum_sq += swap.npv(curve) ** 2
    obj_val.append(sum_sq)
    return sum_sq
```

```python
x0 = [1.0 for i in range(len(pillar_dates)-1)]
bounds = [(0.01, 10.0) for i in range(len(pillar_dates)-1)]
```

```python
from scipy.optimize import minimize
result = minimize(objective_function, x0, bounds=bounds,
                  args=(observation_date, pillar_dates), callback=pippo)

print (result)
```

```python
print ("Initial objective function value ", objective_function(x0, observation_date, 
                                                               pillar_dates))
print ("Final objective function value ", objective_function(result.x, observation_date, 
                                                               pillar_dates))
```

```python
from matplotlib import pyplot as plt

res = result.x
y = []
x = np.arange(0.999, 1.001, 0.00001)
for x1 in x:
    res[0] = x1
    r = objective_function(res, observation_date, pillar_dates)
    y.append(r)

plt.rcParams.update({'font.size': 16, 'figure.figsize':(10, 6)})
from matplotlib.ticker import FormatStrFormatter

sub1 = plt.subplot(1, 2, 1)
sub1.yaxis.set_major_formatter(FormatStrFormatter('%.1e'))
sub1.xaxis.set_major_formatter(FormatStrFormatter('%.4f'))
sub1.plot(x, y)
sub1.plot(0.99990873, 0, marker='o', markersize=12, color='red')
sub1.grid(True)
sub1.set_ylabel('objective function value')
sub1.set_xlabel('$x_1$')
sub1.set_xlim(0.999, 1.001)

sub2 = plt.subplot(1, 2, 2)
sub2.yaxis.set_major_formatter(FormatStrFormatter('%.1e'))
sub2.plot(range(len(obj_val)), obj_val[:100])
#sub2.plot(0.99990873, 0, marker='o', markersize=12, color='red')
sub2.grid(True)
#sub2.set_ylabel('objective function value')
sub2.set_xlabel('iteration')
sub2.set_xlim(0, 8)
sub2.set_yscale('log')
```

```python
from numpy.random import normal, seed
from matplotlib import pyplot as plt
import math
from finmarkets import call

seed(1)
S = 100
mu = 0.0
sigma = 0.05
T=1
historical_series = []
N = 3000
C0 = call(S, 105, 0, 0.05, 1)
payoff = 0
print (C0)
for i in range(1, N):
    S = 100 * math.exp((mu - 0.5 * sigma * sigma) * T + sigma * math.sqrt(T) * normal())
    #print (S)
    payoff += max(0, S - 105)
    historical_series.append((payoff/i - C0)/C0)
print (payoff)
plt.plot(range(1, N), historical_series)
plt.xlabel('days')
plt.ylabel('stock price')
plt.grid(True)
plt.ylim(-0.35, 1)
plt.xlim(0, N)
plt.show()
```
