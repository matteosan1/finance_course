```python
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials
import pandas as pd, time
from random import seed, sample

seed(time.time())
projects = [
'Basis Swap Par Spreads',
'Credit Curve Bootstrapper',
'Credit VAR of a Swap',
'Fixed Coupon Bond',
'Floating Coupon Bond',
'Libor Curve Bootstrapper',
'Swap with Counterparty Risk (CVA)',
'Portfolio Optimization',
'Model Calibration With Machine Learning',
'Cross Currency Swap (XCS)',
'Gaussian Correlation'
]

gc = gspread.authorize(GoogleCredentials.get_application_default())
worksheet = gc.open('2021_groups').sheet1
rows = worksheet.get_all_values()

df = pd.DataFrame.from_records(rows)
df = df[(df.index >= 4) & (df[1] != "")]

assignement = sample(projects, len(df))
for i in range(len(df)):
  print (df.iloc[i, 1:3].values, assignement[i])
```
