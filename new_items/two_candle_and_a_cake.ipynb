```python
import numpy as np

def experiment(e, trials):
  np.random.seed(e)

  succ = 0
  for _ in range(trials):
    x = np.random.uniform(size=3)
    if x[0] < x[1] < x[2] or x[2] < x[1] < x[0]:
      succ += 1

  return (succ/trials)


r = []
for e in range(10000):
  r.append(experiment(e, 10000))
```

```python
import matplotlib.pyplot as plt

plt.hist(r, bins=33, range=(0.315, 0.350))
plt.show()
print (np.mean(r))
print (np.std(r)/np.sqrt(10000))
```
